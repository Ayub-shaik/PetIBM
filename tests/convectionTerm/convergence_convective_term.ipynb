```python
import os

import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
os.system('make cleanConvection')
os.system('make')
```

```python
try:
    os.remove('relative_errors_2d.dat')
except OSError:
    pass
os.system('make runConvection2d')
```

```python
try:
    os.remove('relative_errors_3d.dat')
except OSError:
    pass
os.system('make runConvection3d')
```

```python
with open('relative_errors_2d.dat', 'r') as infile:
    n_cells, relative_errors = numpy.loadtxt(infile, dtype=float, 
                                             unpack=True)

pyplot.figure(figsize=(8, 8))
pyplot.grid(True)
pyplot.xlabel(r'grid-size', fontsize=16)
pyplot.ylabel(r'relative error in convective term', fontsize=16)
pyplot.plot(n_cells, relative_errors, label=r'convective term', 
            color='#D95F02', lw=1, ls='-', marker='o', markersize=6)
n = numpy.linspace(n_cells[0], n_cells[-1], 1001)
pyplot.plot(n, relative_errors[0]*(n_cells[0]/n), 
            label=r'$1^{nd}$-order convergence',
            color='k', lw=1, ls='-')
pyplot.plot(n, relative_errors[0]*(n_cells[0]/n)**2, 
            label=r'$2^{nd}$-order convergence',
            color='k', lw=1, ls='--')
pyplot.legend(loc='best', prop={'size':16}, frameon=False)
pyplot.xscale('log')
pyplot.yscale('log');
```
