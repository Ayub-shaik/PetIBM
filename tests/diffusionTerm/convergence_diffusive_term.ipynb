# Spatial convergence of the explicit diffusive terms

```python
import os

import numpy
from matplotlib import pyplot
%matplotlib inline
```

## Clean and compile the unit-test for the convective term

```python
os.system('make cleanDiffusion')
os.system('make')
```

## Run unit-test on different two-dimensional grids

```python
try:
    os.remove('relative_errors_2d.dat')
except OSError:
    pass
os.system('make runDiffusion2d')
```

## Run unit-test on different three-dimensional grids

```python
try:
    os.remove('relative_errors_3d.dat')
except OSError:
    pass
os.system('make runDiffusion3d')
```

## Plot two-dimensional spatial convergence

```python
with open('relative_errors_2d.dat', 'r') as infile:
    h, relative_errors = numpy.loadtxt(infile, dtype=float, unpack=True)

pyplot.figure(figsize=(8, 8))
pyplot.grid(True)
pyplot.xlabel(r'grid-spacing', fontsize=16)
pyplot.ylabel(r'relative error in diffusive term', fontsize=16)
pyplot.plot(h, relative_errors, 
            label=r'diffusive term', 
            color='#D95F02', lw=1, ls='-', marker='o', markersize=6)
h_gauge = numpy.linspace(h[0], h[-1], 101)
pyplot.plot(h_gauge, relative_errors[0]*h_gauge/h[0], 
            label=r'$1^{nd}$-order convergence',
            color='k', lw=1, ls='-')
pyplot.plot(h_gauge, relative_errors[0]*(h_gauge/h[0])**2, 
            label=r'$2^{nd}$-order convergence',
            color='k', lw=1, ls='--')
pyplot.legend(loc='best', prop={'size':16}, frameon=False)
pyplot.xscale('log')
pyplot.yscale('log');
```

## Plot three-dimensional spatial convergence

```python
with open('relative_errors_3d.dat', 'r') as infile:
    h, relative_errors = numpy.loadtxt(infile, dtype=float, unpack=True)

pyplot.figure(figsize=(8, 8))
pyplot.grid(True)
pyplot.xlabel(r'grid-spacing', fontsize=16)
pyplot.ylabel(r'relative error in diffusive term', fontsize=16)
pyplot.plot(h, relative_errors, 
            label=r'diffusive term', 
            color='#D95F02', lw=1, ls='-', marker='o', markersize=6)
h_gauge = numpy.linspace(h[0], h[-1], 1001)
pyplot.plot(h_gauge, relative_errors[0]*h_gauge/h[0], 
            label=r'$1^{nd}$-order convergence',
            color='k', lw=1, ls='-')
pyplot.plot(h_gauge, relative_errors[0]*(h_gauge/h[0])**2, 
            label=r'$2^{nd}$-order convergence',
            color='k', lw=1, ls='--')
pyplot.legend(loc='best', prop={'size':16}, frameon=False)
pyplot.xscale('log')
pyplot.yscale('log');
```
